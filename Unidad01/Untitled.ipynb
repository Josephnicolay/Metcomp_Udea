```python
import math
import numpy as np
import matplotlib.pyplot as plt
from random import randint
    
```

```python
def harmonic_oscillator(x0,w,t):#función(posición inicial,frecuencia angular,tiempo)
    
    position=[]
    velocity=[]
    aceleration=[]
   
    time=np.arange(0,t,0.01)
    
    
    for i in range(0,len(time)):
               
        position.append(x0*np.cos(w*time[i]))
        velocity.append((-1)*x0*w*np.sin(w*time[i]))
        aceleration.append((-1)*x0*(w**2)*np.cos(w*time[i]))
        
    return position,velocity,aceleration,time
```

```python
x0 = randint(1,15)
w = randint(1,15)
t =10

position,velocity,aceleration,time = harmonic_oscillator(x0,w,t)

plt.figure()

#revisar
plt.subplot(2,2,1)
plt.plot(time,position, '.')

plt.subplot(2,2,2)
plt.plot(time,velocity, '.')


plt.subplot(2,2,3)
plt.plot(time,aceleration, '.')

plt.show()
```
