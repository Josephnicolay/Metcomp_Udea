```python
import math as m
import numpy as np
import matplotlib.pyplot as plt

G=6.67e-11
Ms=1.989e30 #Masa del Sol
AU=146.6e9 #unidades astro. en metros
a=17.834*AU
e=0.96714
year=86400*365.25#conversion los a√±os a segundos 
t_ini=0.0
t_end=75.32*year
time_array = np.linspace(t_ini,t_end,1000)
#definimos la funcion de kepler
def f(E,t):    
    value = E-e*np.sin(E)-(G*Ms/a**3)**0.5*t
    return value
def dfdE(E):
    value = 1-e*np.cos(E)
    return value
```

```python


Nmax=1000
tol=1e-6


phi_array=[]
r_array=[]

for i in range (0,len(time_array)):
    error=1.0
    p_prev=(G*Ms/a**3)**0.5*time_array[i] #Anomalia media
    iteration=0
    while(error>tol):
        p=p_prev - f(p_prev,time_array[i])/dfdE(p_prev)
        error = abs(p-p_prev)
        p_prev=p
        iteration=iteration +1   
    
        if (iteration>=Nmax):
            print("El metodo no converge despues de %d iteraciones"%(Nmax))
            break
    phi=2.0*np.arctan(((1+e)/(1-e))**0.5*np.tan(0.5*p))
    r=a*(1-e**2)/(1+e*np.cos(phi))  
    phi_array.append(phi)
    r_array.append(r)
    
x_array=[]
y_array=[]

for i in range(0,len(r_array)):
    x_array.append(r_array[i]*np.cos(phi_array[i]))
    y_array.append(r_array[i]*np.sin(phi_array[i]))
        
        
fig=plt.figure(figsize=(20,20))
ax=plt.axes()
ax.plot(x_array,y_array)
ax.plot(0,0,'*')
ax.set_aspect('equal')
ax.set_xlabel('x[AU]')
ax.set_ylabel('y[AU]')
plt.show()
```
